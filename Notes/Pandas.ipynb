## Group by and Aggregating

```python
df.groupby('Industry').count() 
```

```python
df.groupby('Industry').min()
```

```python
df.groupby('Industry').sum()
```

## Merge, Join, and Concatenate

```python
data = {
    'FellowshipID': [1001, 1002, 1003, 1004],
    'FirstName': ['Frodo', 'Samwise', 'Gandalf', 'Pippin'],
    'Skills': ['Hiding', 'Gardening', 'Spells', 'Fireworks']
}

# Create the DataFrame
df1 = pd.DataFrame(data)
df1
```

```python
data = {
    'FellowshipID': [1001, 1002, 1006, 1007, 1008],
    'FirstName': ['Frodo', 'Samwise', 'Legolas', 'Elrond', 'Boromir'],
    'Age': [50, 39, 2931, 6520, 51]
}

# Create the second DataFrame
df2 = pd.DataFrame(data)
df2
```

```python
df1.merge(df2) # inner join
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4 # join works better with indexes
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join = 'outer', axis=1)
```
